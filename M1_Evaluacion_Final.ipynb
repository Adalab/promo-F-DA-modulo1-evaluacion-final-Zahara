```python
#Creamos una clase llamada TiendaOnline
class TiendaOnline:
    # Inicializamos los atributos
    def __init__(self, nombre):
        self.nombre = nombre
        # Una lista de diccionarios para almacenar los productos del inventario: "nombre", "precio" y "cantidad".
        # Cada diccionario es un producto: {"nombre": "camisa", "precio": 20, "cantidad": 40}
        self.inventario = []
        # Un diccionario de diccionarios para llevar el registro de los clientes. Claves: "nombre", "email" y "compras".
        # Cada subdiccionario es un cliente: {"cliente1": {"email": "cliente1@email.com", "compras": []}, "cliente2": {"email": "cliente2@email.com", "compras": []}}
        self.clientes = {}
        # Un float para llevar un registro de las ventas totales de la tienda.
        # Comenzamos con 0.
        self.ventas_totales = 0.00

# Metodo_1 Agregar productos. 
    # Definimos el método con 4 parámetros (self para la instancia y 3 más).
    def agregar_producto(self, nombre, precio, cantidad):
        # Comprueba si el producto ya está en el inventario de la tienda.
        for producto in self.inventario:
            # Si el nombre del producto ya está en el inventario.
            if  producto["nombre"] == nombre:
                # Si el producto ya existe, actualiza su precio y cantidad de ese artículo en el inventario.
                producto["precio"] = precio
                producto["cantidad"] = cantidad
                # Muestra un mensaje que indica que ya existía y lo ha actualizado.
                print(f"El producto {nombre} ya estaba en el inventario y ha sido actualizado.")
                break # Para el bucle cuando ya está ese nombre en el inventario.
        # Si no se ha encontrado, el bucle se viene a aquí.
        else:
            # Si no está, lo añade con lo que le indicamos en la llamada.
            self.inventario.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})
            # Hacemos un print por cada artículo.
            print(f"El producto {nombre} se ha añadido al catálogo de productos.")
        # Hacemos que muestre el inventario actualizado.
        #return self.inventario  #Comento esta línea porque me lo imprime dos veces, aunque en el ejercicio de clase está así. 


# Método 2: Ver el inventario.
    # Definimos solo con self ya que no vamos a necesitar más.
    def ver_inventario(self):
        # El siguiente print es innecesario pero "hace bonito". 
        print("A continuación se muestra el inventario de la tienda:")
        # Iniciamos un bucle for usando "producto" como "i". Recorre el inventario elemento a elemento.
        for producto in self.inventario:
            # Por cada iteración del bucle, imprimimos los detalles del producto encontrado.
            print(f"Nombre: {producto['nombre']}, Precio: {producto['precio']}, Cantidad: {producto['cantidad']}")


# Método 3: Buscar un producto.
    # Definimos el método.
    def buscar_producto(self, nombre):
        # Iniciamos un bucle for como en el método anterior.
        for producto in self.inventario:
            # Comprueba si el "nombre" del diccionario es igual al nombre del argumento de la función.
            if producto['nombre'] == nombre:
                # Si coincide, imprime la información del producto.
                print(f"Nombre: {producto['nombre']} Precio: {producto['precio']} Cantidad: {producto['cantidad']}")
                # Detiene la función.
                return      # Si lo quitamos, da la información pero dice que no está en el inventario.  Pregunta 1.
        # Mensaje que sale si el bucle for no encuentra el producto.
        print(f"El producto {nombre} no se encuentra en el inventario.")



# Método 4: Actualizar stock. 
    # Definimos con 3 parámetros.
    def actualizar_stock(self, nombre, cantidad):
        # Recorremos la lista con un bucle for.
        for producto in self.inventario:
            # Verificamos si el valor del "nombre" del diccionario coincide con el de la variable "nombre" del init.
            if producto['nombre'] == nombre:
                # Si lo anterior coincide, cambia la cantidad a la espefificada en la variable de la llamada.
                producto['cantidad'] += cantidad
                # Imprime la frase con los nuevos valores.
                print(f"Stock actualizado para el producto {nombre}. Cantidad disponible: {producto['cantidad']}")
                # Finalizamos el bucle for si se daba la circunstancia anterior (encontrar el producto, etc).
                return
        # Si no se encuentra el producto, informamos con un print.
        print(f"El producto {nombre} no se encuentra en el inventario.")
    
    
    
# Método 5: Eliminar producto.
   # Definimos con dos argumentos.
    def eliminar_producto(self, nombre):
        # Iniciamos bucle for para recorrer el inventario buscando un producto.
        for producto in self.inventario:
            #Si el nombre del producto coincide con la variable "nombre"...
            if producto['nombre'] == nombre:
                # Si coincide, usamos .remove() para eliminar ese producto de nuestro inventario.
                self.inventario.remove(producto)  
                # Informamos de la eliminación.
                print(f"{nombre} ha sido eliminado del inventario.")
                # Usamos return para parar el bucle y que no siga buscando. #Pregunta 1.
                return
            # Si el bucle for no encuentra el "nombre"...
            else:
            # Como el producto no se encontró, muestra un mensaje informativo.
             print(f"No se ha encontrado ningún producto llamado {nombre}.")


# Metodo 6: Calcular el valor del inventario.
    # Definimos solo con self.
    def calcular_valor_inventario(self):
        # Inicializa el valor total de 0, para tenerlo vacío para usarlo "limpio"
        valor_total = 0  
        # Iniciamos un bucle for para recorrer la lista mirando los elementos del inventario.
        for producto in self.inventario:        #quiero meter un sum pero no acierto y así funciona, así que nos conformamos con el "minimo funcional".
            # Calculamos el valor del producto actual multiplicando el precio de producto por la cantidad de productos.
            valor_producto = producto["precio"] * producto["cantidad"]
            # Sumamos lo que valen todas las unidades de ese producto al valor total y así con cada producto (bucle), al final tenemos el de todo el inventario.
            valor_total += valor_producto 
        # Imprimimos el valor total del inventario
        return (f"Nuestro inventario tiene un valor total de {valor_total}€") # Paréntesis no necesario pero no molesta.
    
##Método 7: Buscar un producto que coincida con un patrón de busqueda en formato regex.
#    # Importamos regex para poder usarlo.
#    import re
#    #Definimos con dos argumentos.
#    def buscar_producto_regex (self, patron):
#        #Inicializamos una lista vacía donde almacenar los productos que tengan el patrón en su nombre.
#        productos_coincidentes = []
#        #Iteramos por el inventario.
#        for producto in self.inventario:
#            # Si en el inventario se encuentra algún producto que coincida con el patrón...
#            if re.search(patron, producto["nombre"]):  
#                # Añadimos ese producto a la lista de productos coincidentes.
#                productos_coincidentes.append(producto)
#        # Comprobamos si hay algo en la lista de productos coincidentes.
#        if productos_coincidentes:
#            print("Los productos que coinciden con el patrón de búsqueda son:")
#            # Iteramos por la lista para ir producto por producto.
#            for producto in productos_coincidentes:
#                # Imprimimos el nombre y precio de los productos de la lista.
#                print(f"Nombre: {producto['nombre']}, Precio: {producto['precio']}")
#        # Si no hay productos en la lista...
#        else:
#            # ... mostramos un mensaje informativo.
#            print("En el inventario no hay productos que coincidan con ese patrón.")

            


# Método 8: Realizar una compra.
    # Definimos la función solo con self.
    def realizar_compra(self):
        # Llamamos a ver_inventario para que muestre el cliente sepa lo que tenemos en la tienda. 
        self.ver_inventario()
        # Iniciamos una lista vacía para que haga de cesta.
        cesta = []
        # Mientras la cesta esté vacía, su precio será 0€.
        precio_cesta = 0
        # Iniciamos un bucle que sigue mientras la condición sea True.
        while True:
            # Interactuamos con el usuario y almacenamos la respuesta en una variable.
            quieres_comprar = input("¿Quieres comprar uno de nuestros productos? (S/N)").lower() # Convertimos a minúscula.
            # Comprueba si la respuesta del usuario es diferente a "s".
            if quieres_comprar != "s":
                # Si es distinta, comprobamos si es difernte a "n".
                if quieres_comprar != "n":
                    # Si es diferente a ambas, imprime un mensaje.
                    print("Respuesta no válida. Por favor, ingresa 'S' para comprar o 'N' para salir.")
                    # Seguimos con el bucle hasta que nos introduzca una respuesta correcta.
                    continue
                #Se ejecuta si la respuesta del usuario sí es "n".
                else:
                    # Como el usuario no quiere hacer una compra, paramos el bucle.
                    break

            # No funciona para sustituir lo de arriba. Preguntar a Ana.
            #if quieres_comprar != "s" and quieres_comprar != "n":
            #    print ("Respuesta no válida. Por favor, ingresa 'S' para comprar o 'N' para salir.")
            #    continue
            #    else:
            #         break


            #Si la respuesta no es diferente a "s", es que sí quiere comprar, así que le preguntamos qué quiere comprar (recordemos que está viendo el inventario).
            compra_usuario = input("¿Qué producto te interesa? (Indica el nombre)").lower() # Convertimos a minúsculas.
            #Inicializa una variable None para rastrear si se encuentra un producto o no.
            producto_encontrado = None
            # Con un bucle for recorremos el inventario busando productos.
            for producto in self.inventario:
                # Comprobamos si el nombre de alguno de nuestros productos coincide con el ingresado por el usuario.
                if producto["nombre"].lower() == compra_usuario:  #lower se queda blanco, ¿por qué?
                    # Si encuentra algún producto que coincida, lo alamacena en producto_encontrado.
                    producto_encontrado = producto
                    # Rompe el bucle for.
                    break
            # Comprobamos si se encontró algún producto con ese nombre.
            if producto_encontrado:
                # Si se encontró, solicitamos al usuarioque ingrese cuántas unidades quiere de ese producto. Tiene que ser un integer.
                cantidad_articulos_deseada = int(input(f"¿Cuántas unidades de {compra_usuario} quieres? (Indica la cantidad)"))
                # Comprobamos si la cantidad de unidades que ha indicado es mayor que las que tenemos en el inventario.
                if cantidad_articulos_deseada > producto_encontrado["cantidad"]:
                    # Si es mayor, imprimimos un mensaje informativo.
                    print(f"Solo hay {producto_encontrado['cantidad']} unidades de {compra_usuario} en la tienda.")
                # Si tenemos suficiente cantidad, ejecutamos a partir de aquí.
                else:
                    # Calcula el precio total de la cantidad deseada de ese producto con una multiplicación del precio del producto y la cantidad deseada.
                    precio_cantidad_producto = producto_encontrado["precio"] * cantidad_articulos_deseada
                    # Añade la cantidad obtenida al precio total de la cesta (ya que puede tener articulos anteriormente).
                    precio_cesta += precio_cantidad_producto
                    # Agrega una tupla que contiene el nombre del artículo y la cantidad deseada a la cesta.
                    cesta.append((producto_encontrado["nombre"], cantidad_articulos_deseada))
                    # Informa al usuario sobre lo que hay en su cesta y el precio que tiene. 
                    print(f"Ya tienes {cantidad_articulos_deseada} '{compra_usuario}' en tu cesta.\n El precio de tu cesta es de: {precio_cesta}€")
                    # Restamos la cantidad de producto del inventario.
                    producto_encontrado["cantidad"] -= cantidad_articulos_deseada
            # Si no se encontró ningún artículo que coincida con la respuesta del usuario, ejecuta desde aquí.
            else:
                # Imprime un mensaje informativo.
                print(f"El producto '{compra_usuario}' no consta en el inventario.")
        # Comprobamos que hay algo en la cesta.
        if cesta:
            # Imprimimos un mensaje introductorio al resumen de la compra.
            print("Este es su pedido:")
            # Con un bucle for recorremos toda la cesta.
            for producto, cantidad in cesta:
                # Mostramos un mensaje con el nombre de productos y la cantidad de cada uno que hay en la cesta.
                print(f"{cantidad} '{producto}'")
            # Mostramos el precio total de la cesta con todos los productos.
            print(f"El costo total de la compra asciende a {precio_cesta}€")
        #Si no había nada en la cesta... 
        else:
            # ... mostramos un mensaje informativo.
            print("No se ha agregado ningún producto al carrito.")



# Método 9: Procesar el pago de una compra.
    # Definimos con un solo argumento.
    def procesar_pago(self):
        # Empezamos un código de manejar excepciones. Se intenta el try y si hay algún error, lo manejamos con except. 
        try:
            # Solicitamos al cliente ingresar la cantidad total a pagar. 
            total_a_pagar = float(input("Por favor, ingrese el precio de su compra "))
            # Solicitamos al cliente ingresar la cantidad que proporciona.
            cantidad_pago = float(input("Por favor, ingrese el pago que hace para poder calcular su cambio."))
            # Calculamos el cambio restando al dinero que da lo que debe pagar.
            cambio = cantidad_pago - total_a_pagar
            # Verificar si el pago es mayor que el total a pagar. Si el cambio es mayor o igual a cero, significa que el pago es suficiente.
            if cambio >= 0:
                # Sumamos esta venta a las ventas totales de la tienda.
                self.ventas_totales += total_a_pagar
                # Imprimimos un mensaje infomando del cambio.
                print(f"Su cambio es de {cambio:.2f}€. ¡Vuelva pronto!") #lo de .2F venía en google, no sé cómo va pero funciona y no sé otra manera.
            # Si el dinero proporcionado por el cliente no es suficiente(menor que el total a pagar), se ejecuta else.
            else:
                # Mostramos mensaje informativo.
                print("Cantidad insuficiente. No puede realizar el pago con esa cantidad.") #No sé si debería de seguir ejecutándose el input.
        # Manejar cualquier error como dar letras o símbolos en vez de números.
        except ValueError:
            # Mostramos mensaje informando del error.
            print("Error: Ingrese valores numéricos válidos para la cantidad total y la cantidad de pago.")


# Método 10: Agregar un cliente.   #OJO, en el ejercicio de clase eran listas y quiero diccionarios.
    # Definimos con 3 argumentos.
    def agregar_cliente(self, nombre, email):
        # Comprobamos si el nombre del cliente no está en el diccionario de clientes.
        if nombre not in self.clientes:
            # Si no existe, añadimos el cliente al diccionario de clientes con su nombre e email (y un diccionario vacío "compras" para el ejercicio 12).
            self.clientes[nombre] = {"email": email, "compras": []}
            # Imprimimos un mensaje confirmando que hemos añadido ese cliente.
            print(f"Cliente '{nombre}' añadido a la lista de clientes.")
        # Si el nombre del cliente sí estaba en el diccionario...
        else:
            # ... mostramos un mensaje informativo.
            print(f"El cliente {nombre} ya existe en la base de datos.")


# Método 11: Ver clientes existentes.
    # Definimos con un solo argumento.
    def ver_clientes(self):
        # Creamos una lista vacía para la información que tenemos sobre los clientes.
        clientes_info = []
        # Iteramos por la lista de clientes.
        for nombre, email in self.clientes.items():
            # Apendeamos la información básica a la lista recién creada como una tupla (nombre, email).
            clientes_info.append((nombre, email))
        # Devuelve la lista con la información de los clientes.
        return f"La lista de cientes se muestra a continuación: {clientes_info}"    #Poner bonito el "print" que está feo.

# Método 12: Registrar una compra.
    # Definimos con 3 argumentos.
    def registrar_compra(self, nombre_cliente, carrito):
        # Buscamos al cliente en el diccionario de clientes.
        cliente = self.clientes.get(nombre_cliente)
        # Si el cliente está en el diccionario... 
        if cliente:
            # Calculamos el total de la compra multiplicando el precio del producto por la cantidad que quiere el cliente y eso lo sumamos a total_compra..
            total_compra = sum(brocoli["precio"] * brocoli["cantidad"] for brocoli in carrito.values())
            # Añadimos esta compra en la lista de compras del cliente.
            cliente["compras"].append({"carrito": carrito, "total": total_compra})  #¿Por qué append sigue en blanco? ¿Y values?
            # Añadimos esta compra a las ventas totales de la tienda.
            self.ventas_totales += total_compra
            # Imprimimos un mensaje informativo de que se ha añadido a su historial.
            print(f"La compra del cliente '{nombre_cliente}'se ha añadido a su historial. Total de compra: {total_compra}€")
        # Si el cliente no estaba en el diccionario...
        else:
            # ... sale un mensaje informativo de que no se ha podido hacer.
            print(f"Compra no registrada. El cliente '{nombre_cliente}' no se encuentra en el diccionario de clientes.")

   
        
# Método 13: Ver las compras de un cliente.
    # Definimos con dos argumentos.
    def ver_compras_cliente(self, nombre_cliente):
        # Buscamos al cliente por su nombre en el diccionario.
        cliente = self.clientes.get(nombre_cliente)
        # Si el cliente está en el diccionario...
        if cliente:
            # Mostramos una introducción informando de lo que vamos a mostrar abajo.
            print(f"Historial de compras para el cliente '{nombre_cliente}':")
            # Comprobamos que el cliente tenga algo en el historial de compras.
            if cliente["compras"]:
                # Iteramos por cada elemento de su lista de compras. Comenzamos desde el 1 aunque debería ser un 0, pero con 1 funciona bien.
                for brocoli, compra in enumerate(cliente["compras"], 1): #Ponemos 1 en vez de 0 porque entonces miestra la compra 0.
                    # Imprimimos el número de compra, 0-1 como comentábamos.
                    print(f"Compra {brocoli}:")
                    # Iteramos por los elementos de esa compra.
                    for producto, cantidad in compra["carrito"].items():
                        # Imprimimos el nombre del producto y la cantidad que se llevó.
                        print(f"Producto: {producto}, Cantidad: {cantidad}")
                    # Mostramos el importe de la compra entera.
                    print(f"Total de la compra: {compra['total']}€")
                    # Imprimimos rayas a modo de separación para hacer más legible el mensaje.
                    print("-------------------")
            # Si el cliente está en el diccionario no hay nada en el historial de compras del cliente...
            else:
                # Imprimimos un mensaje informativo.
                print(f"{nombre_cliente} no tiene historial de compras.")
                # Imprimimos rayas a modo de separación para hacer más legible el mensaje.
                print("-------------------")
        # Si el cleinte no se encuentra en el diccionario...
        else: 
            # Imprimimos un mensaje informativo.
            print(f"No se puede consultar el historial de compras de '{nombre_cliente}' ya que no está entre nuestros clientes.")
            


#Método 14: Calcular las ventas totales de la tienda.
    # Definimos solo con self.
    def calcular_ventas_totales(self):
        # Inicializamos la variable ventas_totales en 0.
        ventas_totales = 0  
        # Recorremos el historial de compras de todos los clientes cogiendo los valores.
        for cliente in self.clientes.values():
            #Por cada compra que encontremos en cada cliente...
            for compra in cliente["compras"]:
                # Sumamos el importe de cada compra a ventas_totales.
                ventas_totales += compra["total"]
        # Añadimos el total de ventas al atributo self.
        ventas_totales += self.ventas_totales
        # Mostramos el total de ventas de la tienda (ventas no son ganancias).
        return f"El importe total que ha vendido la tienda es de {ventas_totales}€"

```
